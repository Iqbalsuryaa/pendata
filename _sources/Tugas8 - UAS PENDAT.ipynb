Merubah Gender menjadi Numerik

Change Values to 1/0 Take all values of ‘gender’

If values is Female change to 1

If values is Male to 0

```python
# gender code
Gen_wanita = "Female"
Gen_pria = "Male"

# binary value
value_one = 1
value_zero = 0

def change_gender_to_binary(gender):
    return value_one if gender == Gen_pria else value_zero
```

Meng-update hasil perubahah

```python
# Update all values
genderv7["gender"] = genderv7["gender"].apply(change_gender_to_binary)
```

## **>> Pengujian Naive Bayes**

```python
genderv7 =genderv7
X = genderv7.iloc[:,:6].values
y = genderv7['long_hair'].values
genderv7.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
X_std=StandardScaler().fit_transform(X)
X_std
```

```python
df_std = pd.DataFrame(X_std, columns=genderv7.columns[:6])
df_std.to_csv('preprocessed_data.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.05, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

```python
# Mendefinisikan fungsi untuk memprediksi RiskLevel dari input pengguna
def predict_risk_level(input_data):
    # Lakukan preprocessing pada input data
    input_data = sc.transform([input_data])  # Lakukan penskalaan pada input data

    # Lakukan prediksi menggunakan model Gaussian Naive Bayes
    predicted_risk_level = classifier.predict(input_data)

    return predicted_risk_level
```

```python
# Contoh penggunaan fungsi predict_risk_level dengan input pengguna
input_user = [15,120,80,7.01,98,70]
predicted_risk = predict_risk_level(input_user)
print("Predicted Risk Level:", predicted_risk)
```
